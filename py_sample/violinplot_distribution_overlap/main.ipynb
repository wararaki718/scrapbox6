```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data1 = [1, 2, 3, 4, 5]
data2 = [3, 4, 5, 6, 7]
df = pd.DataFrame({'data1': data1, 'data2': data2})
```

```python
sns.violinplot(data=df, palette='Set2')
plt.show()
```

```python
df = pd.DataFrame({
    "category": ["fruit", "sport", "fruit", "sport", "fruit", "sport"],
    "age": [10, 20, 30, 40, 50, 60],
    "purchase": [True, True, True, False, False, False]
})
```

```python
sns.violinplot(data=df, x="age", y="category", hue="purchase", split=True, inner="quart")
plt.show()
```
